```python
gradient = np.linspace(0, 1, 100)
gradient = np.vstack((gradient, gradient))
# print(gradient)


plt.imshow(gradient, cmap = plt.cm.gray_r)
plt.axis('off')
plt.show()
```

```python
gradient = np.linspace(0, 1, 100)
gradient = np.vstack((gradient, gradient))
# print(gradient)


plt.imshow(gradient, cmap = plt.cm.magma)
plt.axis('off')
plt.show()
```

```python
# matplotlib lesson

# first way
plt.figure(figsize=(8, 4))

plt.plot([1, 2, 3, 4, 5, 3])


plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4, 5, 3])

plt.subplot(2, 2, 2)
plt.plot([100, 111, 3123, 123123123])

plt.subplot(2, 2, 3)
plt.plot([1, 2, 3, 4, 5, 3])


plt.subplot(2, 2, 4)
plt.plot([1, 2, 3, 4, 5, 3])

plt.subplot(3, 3, 5)
plt.plot([6, 5, 4, 3, 2] )
plt.plot([1, 2,34, 5, 5] )

plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(3, 3, 1)
plt.plot([1, 2, 3, 4, 5, 3])

plt.subplot(3, 3, 2)
plt.plot([100, 111, 3123, 123123123])

plt.subplot(3, 3, 3)
plt.plot([1, 2, 3, 4, 5, 3])


plt.subplot(3, 3, 4)
plt.plot([1, 2, 3, 4, 5, 3])

plt.subplot(3, 3, 9)
plt.plot([1, 2, 3, 4, 5, 3])

plt.show()
```
