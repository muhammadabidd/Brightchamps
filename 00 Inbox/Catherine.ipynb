```python
all_students_data = {}

# only for making the function, not calling it
def add_student():
    name = input("Enter student name: ")
    age = int(input("Enter student age: "))
    grade = input("Enter student grade: ")
    id = len(all_students_data) + 1

    all_students_data[id] = {
        "NAMA" : name,
        "UMUR" : age,
        "KELAS" : grade
    }
    print(f"Student {name} added with ID: {id}")




def print_All_detail(id):
    student = all_students_data.get(id)
    print(f"The student name is : {student['NAMA']}")
    print(f"The student Age is : {student['UMUR']}")
    print(f"The student Grade is : {student['KELAS']}")
    print("--------------------------------")


def remove_student(id):
    if id in all_students_data:

        del all_students_data[id]
        print(f"The student with ID : {id}  has been removed")
    else:
        print("No student with that ID")
```

```python
# we can call function in this cell
add_student()
add_student()

print_All_detail(1)
print_All_detail(2)


remove_student(1)
all_students_data
```
