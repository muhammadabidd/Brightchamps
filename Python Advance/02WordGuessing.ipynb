```python
import random

print("Hello everyone, letsplay word guessing game")

name = input("Please input your name : ")
list_of_word = ['ice', 'ayam', 'kucing', 'kerasakti']
index = random.randint(0, len(list_of_word))

the_word = list_of_word[index]
guessed_letter = ''



while True:


    tebakan = input("insert your tebakan")
    guessed_letter = guessed_letter + tebakan
    # print(guessed_letter)



    wrong = 0
    for letter in the_word:
        if letter in guessed_letter:
            print(letter)
        else:
            print("-")

            wrong = 1

    print(wrong)
    print('############')
        # print(letter)



    if wrong == 0:
        break;


print(f"Congratulation {name}, The correct word is : {the_word}")
```

```python
import random


list_of_word = ['Abid', 'Budi', "joni"]

index = random.randint(0, len(list_of_word) - 1)



theword = list_of_word[index]

myguess = ''

while True:
    input_guess = input('Insert your letter')
    myguess = myguess + input_guess


    complete = True

    for letter in theword:
        if letter in myguess:
            print(letter)
        else:
            print("_")
            complete = False



    if complete:
        break


    print('=' * 100)
    print(complete)




print("CONGRATULATION!")

        


```
